```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the 'votacao_parlamentar.csv' file
df = pd.read_csv('../data/csv/votacao_parlamentar.csv')

df
```

```python
# Convert the 'data' column to datetime
df['data'] = pd.to_datetime(df['data'], format='%Y-%m-%d')

df
```

```python
# Create a new column for the year of the vote
df['year'] = df['data'].dt.year

# Filter the data to include only years between 2003 and 2023
filtered_df = df[(df['year'] >= 2003) & (df['year'] <= 2023)]

filtered_df
```

```python
# Count the number of unique propositions (votes) per year
propositions_per_year = filtered_df.groupby('year')['id_votacao'].nunique().reset_index()

propositions_per_year
```

```python
# Filter data for the specified year range (2004-2023)
propositions_per_year = propositions_per_year[(propositions_per_year['year'] >= 2004) & 
                                              (propositions_per_year['year'] <= 2023)]

# Plot a bar chart for the specified range
plt.figure(figsize=(12, 6))
bars = plt.bar(propositions_per_year['year'], propositions_per_year['id_votacao'], width=0.6)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Propositions', fontsize=14)
plt.title('Yearly Number of Propositions Analyzed (2004-2023)', fontsize=16)
plt.xticks(ticks=propositions_per_year['year'], rotation=45)

# Add values on top of each bar
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # Position horizontally at the center of the bar
        height,                            # Position vertically at the top of the bar
        f'{height}',                       # Display the value
        ha='center', va='bottom', fontsize=10  # Center alignment and font size
    )

plt.tight_layout()
plt.show()
```

```python
# 1. Display the total number of rows (votes) and unique propositions in the dataset
total_votes = filtered_df.shape[0]
unique_propositions = filtered_df['id_votacao'].nunique()

# 2. Display the range of years in the dataset
min_year = filtered_df['year'].min()
max_year = filtered_df['year'].max()

# 3. Display the most frequent political parties appearing in the dataset
top_parties = filtered_df['sigla_partido'].value_counts().head(10)

# 4. Count the number of votes per year
votes_per_year = filtered_df['year'].value_counts().sort_index()

# 5. Count the number of unique propositions per year
propositions_per_year = filtered_df.groupby('year')['id_votacao'].nunique()

# Print the results
print(f"Total number of votes in the dataset: {total_votes}")
print(f"Total number of unique propositions: {unique_propositions}")
print(f"Year range: {min_year} to {max_year}")
print("\nTop 10 most frequent political parties:")
print(top_parties)
print("\nNumber of votes per year:")
print(votes_per_year)
print("\nNumber of unique propositions per year:")
print(propositions_per_year)
```
